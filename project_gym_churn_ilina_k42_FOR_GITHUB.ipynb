### Контекст и задача

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных для снижения оттока. Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Наша задача — провести анализ и подготовить план действий по удержанию клиентов.

А именно:
* научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
* сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
* проанализировать основные признаки, наиболее сильно влияющие на отток;
* сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
    * выделить целевые группы клиентов;
    * предложить меры по снижению оттока;
    * определить другие особенности взаимодействия с клиентами.

<b>Данные клиента за предыдущий до проверки факта оттока месяц:</b><br>
* `gender` — пол;
* `Near_Location` — проживание или работа в районе, где находится фитнес-центр;
* `Partner` — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
* `Promo_friends` — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
* `Phone` — наличие контактного телефона;
* `Age` — возраст;
* `Lifetime` — время с момента первого обращения в фитнес-центр (в месяцах).

<b>Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:</b><br>
* `Contract_period` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
* `Month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах);
* `Group_visits` — факт посещения групповых занятий;
* `Avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента;
* `Avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц;
* `Avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

## Построение модели прогнозирования оттока клиентов

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце:
* Разобьем данные на обучающую и валидационную выборку функцией train_test_split().
* Обучим модель на train-выборке двумя способами:
    * логистической регрессией,
    * случайным лесом.
* Оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке. Сравните по ним модели. Какая модель показала себя лучше на основании метрик?


### Разделение данных, обучение логистическо регрессией и случайным лесом

```python
#разделим наши данные на признаки (матрица X) и целевую переменную (y)

X = df_clean.drop(columns='churn')#ваш код здесь

y = df_clean['churn'] #ваш код здесь

#разделяем модель на обучающую и валидационную выборки

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
# обучим модель на train-выборке логистической регрессией

lr_model = LogisticRegression(solver='liblinear', random_state=0) 

lr_model.fit(X_train, y_train)

lr_predictions = lr_model.predict(X_test)

lr_probabilities = lr_model.predict_proba(X_test)[:,1]
```

```python
# обучим модель на train-выборке случайным лесом
rfc_model = RandomForestClassifier(random_state=0) 

rfc_model.fit(X_train, y_train)

rfc_predictions = rfc_model.predict(X_test)

rfc_probabilities = rfc_model.predict_proba(X_test)[:,1]
```

### Метрики `accuracy`, `precision` и `recall`: сравнение
Оценим метрики `accuracy`, `precision` и `recall` для обеих моделей на валидационной выборке.

```python
# определим функцию, которая будет выводить наши метрики
def print_all_metrics(y_true, y_pred, title='Метрики классификации'):
    cprint(title, 'cyan', attrs=['bold'])
    print('\tДоля верных - Accuracy: {:.3f}'.format(accuracy_score(y_true, y_pred)))
    print('\tТочность - Precision: {:.3f}'.format(precision_score(y_true, y_pred)))
    print('\tПолнота - Recall: {:.3f}'.format(recall_score(y_true, y_pred)))
```

```python
list_pred = {'LogisticRegression':lr_predictions, 'RandomForestClassifier':rfc_predictions}
```

```python
for key,value in list_pred.items():
    print_all_metrics(y_test, value, title=f"Метрики классификациия для {str(key)}:")
```

### Промежуточный вывод
Модель `RandomForestClassifier` показала лучшие результаты, чем `LogisticRegression`, по всем трем метркиам.

Порассуждаем, какая метрика нам важнее.
`Recall` в данном случае показывает, насколько ПОЛНО и без пропусков модель выявляет тех, кто попадет в отток.
`Precision` показывает, какая часть тех клиентов, для которых был спрогнозирвоан отток, действительно в него попала.
`Accuracy` - общая доля правильных ответов.

Нам кажется, что в данном случае важнее именно полнота. Лучше выявить полностью всех, кому грозит отток, и попытаться маркетинговыми мерами их удержать, захватив при этом часть тех, кто уходить от нас не собирался (в этом не усматривается негативных последствий), чем пропустить долю клиентов, которым грозит отток, и окончательно их упустить. 

Иными словами, по нашему мнению, здесь лучше "перебдеть".

## Кластеризация клиентов
Отложим в сторону столбец с оттоком и проведем **кластеризацию** объектов (клиентов).
* Стандартизируем данные.
* Построим матрицу расстояний функцией `linkage()` на стандартизованной матрице признаков и нарисуем дендрограмму. Предположим, какое количество кластеров можно выделить.
* Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. 
* Посмотрим на средние значения признаков для кластеров. Можно ли сразу что-то заметить?
* Построим распределения признаков для кластеров. Можно ли что-то заметить по ним?
* Для каждого полученного кластера посчитаем долю оттока. Отличаются ли они по доле оттока? Какие кластеры склонны к оттоку, а какие — надёжны?

```python
# уберем из датафрейма целевую колонку с показателями оттока
df_clust = df_clean.drop(columns='churn', axis=1)
df_clust.head()
```

### Стандартизация и дендрограмма

```python
# стандартизация данных 
sc = StandardScaler()
X_sc = sc.fit_transform(df_clust)

linked = linkage(X_sc, method = 'ward') 

# визуализируем как дендрограмму
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация клиентов фитнес-центра')
plt.xticks([])
plt.show() 
```

Визуально можно выделить 4-6 групп (4 - если по цветам, предложенным алгоритмом, 6 - если взять расстояние между кластерамина уровне +/- 50).

### Обучаем модель кластеризации

Обучим модель кластеризации на основании алгоритма K-Means (возьмем число кластеров `n=5`) и спрогнозируем кластеры клиентов.

```python
km = KMeans(n_clusters=5, random_state=0) 
labels = km.fit_predict(X_sc) 
```

```python
df_clean['cluster_km'] = labels
```

Посмотрим на средние значения признаков для кластеров.

* Наибольшие доли оттока в кластерах 1 и 2 - до половины записей, отнесенных к этим кластерам.

* Показатель `current_activity_coef` самый низкий в кластере 0, где наиболее высок уровень оттока, там также самый низкий уровень использованных месяцев. Подтверждается наше предположение о том, что оттекают в основном новички, не успевшие распробовать прелесть физ. активности.
